# A little bit of linear algebra

## Eigenvalues ​​and eigenvectors

Given a square matrix $\mathbf{A} \in \mathbb{R}^{n\times n}$, if scalar $\lambda$ and vector $\mathbf{x}\neq 0\in \mathbb{R}^n$ satisfy:

$$\mathbf{Ax} = \lambda \mathbf{x}$$

then $\lambda$ is called an eigenvalue of $\mathbf{A}$ and $\mathbf{x}$ is called an eigenvector corresponding to that eigenvalue.

Several properties:

- If $\mathbf{x}$ is a eigenvector of $\mathbf{A}$, $\alpha \mathbf{x}$ is also a eigenvector of $\mathbf{A}$ for any scalar $\alpha \neq 0$.
- Every square matrix of order $n$ has $n$ eigenvalues ​​(including repetitions) and can be complex numbers. 
- With symmetric matrices, all eigenvalues are real numbers.
- With [positive definite matrices](https://rdsic.edu.vn/blog/toan/ma-tran-xac-dinh-duong-nhung-ly-thuyet-co-ban-va-ung-dung-vi-cb.html), all its eigenvalues are positive real numbers. With [positive semidefinite matrices](https://rdsic.edu.vn/blog/toan/meo-on-tap-ma-tran-nua-xac-dinh-duong-tong-hop-vi-cb.html), all its eigenvalues are non-negative real numbers.

The last property can be inferred from the definition of positive (semi)definite matrices. Indeed, let $\mathbf{u} \neq \mathbf{0}$ be eigenvectors corresponding to the eigenvalue $\lambda$ of positive definite matrix $\mathbf{A}$, we have:

$$\mathbf{Au} = \lambda \mathbf{u} \Rightarrow \mathbf{u}^T\mathbf{Au} = \lambda \mathbf{u}^T\mathbf{u} = \lambda ||\mathbf{u}||_2^2$$

because $\mathbf{A}$ is semidefinite positive, for all $\mathbf{u} \neq \mathbf{0}, \mathbf{u}^T\mathbf{Au} \geq 0$ infer $||\mathbf{u}||_2^2 > 0$. Therefore, we obtain $\lambda > 0$.

## Orthogonal and orthonormal systems

A base system ${\mathbf{u}_1, \mathbf{u}_2,\dots, \mathbf{u}_m \in \mathbb{R}^m}$ is called orthogonal if each vector is different from $0$ and the product of two arbitrary distinct vectors is equal to $0$:

$$\mathbf{u}_i \neq \mathbf{0}; ~~ \mathbf{u}_i^T \mathbf{u}_j = 0, ~ \forall i\neq j \in [1, m]$$

A base system ${\mathbf{u}_1, \mathbf{u}_2,\dots, \mathbf{u}_m \in \mathbb{R}^m}$ is call orthonormal if it is orthogonal and the Euclidean norm of each vector is equal to $1$:

$$
\begin{align}
\mathbf{u}_i^T \mathbf{u}_j = \left\{
\begin{matrix}
    1 & \text{if} &i = j \newline
    0 & \text{otherwise}
\end{matrix}
\right.
\end{align}
$$

Let $\mathbf{U} = [\mathbf{u}_1, \mathbf{u}_2,\dots, \mathbf{u}_m] \in \mathbb{R}^{n\times m}$ with $\mathbf{u}_i$ are orthonormal vectors. From $(1)$, we have:

$$\mathbf{UU}^T = \mathbf{U}^T\mathbf{U} = \mathbf{I}$$

where $\mathbf{I}$ is the identity matrix of order $m$. We call $\mathbf{U}$ an orthogonal matrix. This type of matrix is not called orthonormal because the definition of orthonormal matrix is not exists.

There are several properties:

- $\mathbf{U}^T = \mathbf{U}^{-1}$: the inverse of an orthogonal matrix is its transpose.
- If $\mathbf{U}$ is orthogonal, $\mathbf{U}^T$ is also orthogonal.
- The determinant of orthogonal matrix is $1$ or $-1$. This can be inferred from $\det(\mathbf{U}) = \det(\mathbf{U}^T)$ and $\det(\mathbf{UU}^T) = \det(\mathbf{U})\det(\mathbf{U}^T) = \det(\mathbf{I}) = 1$.
- Orthogonal matrix represents rotation of a vector. Suppose that we have two vectors $\mathbf{x, y} \in \mathbb{R}^m$ and $\mathbf{U} \in \mathbb{R}^{m\times m}$ is an orthogonal matrix. Using this matrix to rotate two vectors above, we obtain $\mathbf{Ux}$ and $\mathbf{Uy}$. The scalar product of two new vectors is

$$(\mathbf{Ux})^T (\mathbf{Uy}) = \mathbf{x}^T \mathbf{U}^T \mathbf{Uy} = \mathbf{x}^T\mathbf{y}$$

so rotation does not change the dot product between two vectors.
- Suppose that $\hat{\mathbf{U}} \in \mathbb{R}^{m\times r}$ is a submatrix of orthogonal matrix $\mathbf{U}$ created by selecting $r$ columns of $\mathbf{U}$. Therefore, we have $\hat{\mathbf{U}}^T\hat{\mathbf{U}} = \mathbf{I}$, this can be inferred from the definition $(1)$.

# Introduction to SVD

Matrix diagonalization problem says that a square matrix $\mathbf{A} \in \mathbb{R}^{n\times n}$ is called diagonalizable if there exists a diagonal matrix $\mathbf{D}$ and an invertible matrix $\mathbf{P}$ such that:

$$\mathbf{A} = \mathbf{P} \mathbf{D} \mathbf{P}^{-1} \tag{2}$$

The number of non-zero elements of the diagonal matrix $\mathbf{D}$ is the rank of matrix $\mathbf{A}$. Multiplying both sides of $(2)$ by $\mathbf{P}$ we have:

$$\mathbf{AP} = \mathbf{PD} \tag{3}$$

Let $\mathbf{p}_i$ and $\mathbf{d}_i$ be the i-th columns of matrices $\mathbf{P}$ and $\mathbf{D}$, respectively. Since each column of the left-hand side and the right-hand side of $(3)$ must be equal (i.e. in the product of two matrices, the column of the resulting matrix is ​​determined by the right matrix),  we need to have

$$\mathbf{Ap}_i = \mathbf{Pd}_i = d_{ii}\mathbf{p}_i \tag{4}$$

with $d_{ii}$ is the i-th diagonal element of matrix $\mathbf{D}$. The second equality occurs because $\mathbf{D}$ is a diagonal matrix, i.e., the $d_{ii}$ component is nonzero. Expression $(4)$ shows that each element $d_{ii}$ must be an eigenvalue of A and each column vector $\mathbf{p}_i$ must be an eigenvector of A corresponding to the eigenvalue $d_{ii}$.

The decomposition of a square matrix into factors such as $(1)$ is also known as Eigen Decomposition. An important point is that this decomposition only applies to square matrices and does not always exist. It exists only if the matrix $\mathbf{A}$ has $n$ linearly independent eigenvectors, since otherwise there is no invertible matrix $\mathbf{P}$ (i.e. if there is no exists $n$ linearly independent eigenvectors, then the set $S = \{\mathbf{p}_i\}$ will be linear dependence. Accordingly, we deduce $\det(P) = 0$ so $\mathbf{P}$ is not invertible). Furthermore, this decomposition is not unique because if $\mathbf{P, D}$ satisfy $(1)$, then $k\mathbf{P, D}$ also satisfy for any nonzero real number $k$.

## Singular Value Decomposition statement

I will write $\mathbf{A}_{m\times n}$ to denote a matrix $\mathbf{A}\in \mathbb{R}^{m\times n}$. Any matrix $\mathbf{A}_{m\times n}$ can be decomposed into the form:

$$\mathbf{A}_{m \times n} = \mathbf{U}_{m \times m}\mathbf{\Sigma}_{m \times n} (\mathbf{V}_{n \times n})^T \tag{5}$$

In which, $\mathbf{U}$ and $\mathbf{V}$ are orthogonal matrices and $\mathbf{\Sigma}$ is a diagonal matrix with the similar size of $\mathbf{A}$. The elements on the main diagonal of $\mathbf{\Sigma}$ are nonnegative and are in decreasing order $\sigma_1 \geq \sigma_2 \geq \dots \geq\sigma_r \geq 0 = 0 = \dots = 0$. The number of non-zero elements of $\mathbf{\Sigma}$ is the rank of matrix $\mathbf{A}: r = \text{rank}(\mathbf{A})$. Note that although $\mathbf{\Sigma}$ is not a square matrix, we still consider it as a diagonal matrix if its non-zero components are on the main diagonal.

If you are confused about the existence of the SVD, you can refer to the proof in [here](https://courses.cs.duke.edu//cps111/spring07/notes/12.pdf). In addition, the presentation of the SVD in $(5)$ is not unique, because if we just adjust the signs of $\mathbf{U}$ and $\mathbf{V}$ then $(5)$ still holds.

## Origin of the name SVD

From $(5)$, we have:

$$
\begin{align*}
\mathbf{AA}^T &= \mathbf{U}\mathbf{\Sigma} \mathbf{V}^T (\mathbf{U}\mathbf{\Sigma} \mathbf{V}^T)^T \newline
&= \mathbf{U}\mathbf{\Sigma} \mathbf{V}^T \mathbf{V}\mathbf{\Sigma}^T\mathbf{U}^T \newline
&= \mathbf{U}\mathbf{\Sigma}\mathbf{\Sigma}^T\mathbf{U}^T =  \mathbf{U}\mathbf{\Sigma}\mathbf{\Sigma}^T\mathbf{U}^{-1} \tag{6}
\end{align*}
$$

We can observe that $\mathbf{\Sigma \Sigma}^T$ is a diagonal matrix with the elements on the main diagonal being $\sigma_1^2, \sigma_2^2, \dots$ Thus $(6)$ is an eigen decomposition of $\mathbf{AA}^T$. Furthermore, $\sigma_1^2, \sigma_2^2, \dots$ are eigenvalues ​​of $\mathbf{AA}^T$. The matrix $\mathbf{AA}^T$ is always positive semidefinite, so its eigenvalues ​​are nonnegative. The $\sigma_i$, which are the square roots of the eigenvalues ​​of $\mathbf{AA}^T$, are also called the singular values ​​of $\mathbf{A}$. This is where the name singular value decomposition comes from.

Accordingly, each column of $\mathbf{U}$ is an eigenvector of $\mathbf{AA}^T$. We call each of these columns a left-singular vector of $\mathbf{A}$. Similarly, $\mathbf{A}^T \mathbf{A} = \mathbf{V \Sigma}^T \mathbf{\Sigma V}^T$ and the columns of $\mathbf{V}$ are also called right-singular vectors of $\mathbf{A}$.

We will use Python to calculate the SVD of a matrix:

```python
import numpy as np
from numpy import linalg

m, n = 2, 3
A = np.random.rand(m, n)

U, S, V = linalg.svd(A)
# A = U*S*V (variable V returns V^T)
# S is just returned the diagonal elements of the matrix 

# checking if U, V are orthogonal and S is a diagonal matrix with nonnegative decreasing elements
print('Frobenius norm of (UU^T - I) =', linalg.norm(U.dot(U.T) - np.eye(m)))
print('S = ', S)
print('Frobenius norm of (VV^T - I) =', linalg.norm(V.dot(V.T) - np.eye(n)))
```

## Several extensive concepts of SVD

### Singular value of a positive semi-definite matrix

Suppose $\mathbf{A}$ is a positive semi-definite square symmetric matrix, we will prove that the singular values ​​of $\mathbf{A}$ are its own eigenvalues. Indeed, let $\lambda$ be an eigenvalue of $\mathbf{A}$ and $\mathbf{x}$ be an eigenvector corresponding to that eigenvalue, furthermore $||\mathbf{x}||_2 = 1$. Since $\mathbf{A}$ is positive semidefinite, we must have $\lambda \geq 0$. We have

$$\mathbf{Ax} = \lambda \mathbf{x} \Rightarrow \mathbf{A}^T \mathbf{A}\mathbf{x} = \lambda \mathbf{A}\mathbf{x} = \lambda^2 \mathbf{x}$$

Note that the positive semi-definite matrix $\mathbf{A}$ has an important property that $\mathbf{A}$ is a symmetric matrix, meaning that $\mathbf{A}^T = \mathbf{A}$. Therefore, $\lambda^2$ is an eigenvalue of $\mathbf{A}^T \mathbf{A}$ so singular value $\mathbf{A}$ is $\sqrt{\lambda^2} = \lambda$.

### Compact SVD

Rewrite the expression $(5)$ under the form of a sum of rank-1 matrices:

$$\mathbf{A} = \sigma_1 \mathbf{u}_1 \mathbf{v}^T_1 + \sigma_2\mathbf{u}_2\mathbf{v}_2^2 + \dots + \sigma_r\mathbf{u}_r\mathbf{v}_r^T$$

with note that each $\mathbf{u}_i \mathbf{v}^T_i, 1 \leq i \leq r$ is a rank-1 matrix.

It is clear that in this representation, the matrix $\mathbf{A}$ depends only on the first $r$ columns of $\mathbf{U, V}$ where $r$ is defined as the $r$ non-zero values ​​of the diagonal matrix $\mathbf{\Sigma}$. This is called the compact SVD: 

$$\mathbf{A} = {\mathbf{U}}_r{\Sigma}_r({\mathbf{V}}_r)^T$$

where $\mathbf{U}_r$ and $\mathbf{V}_r$ are the matrices formed by the first $r$ columns of $\mathbf{U}$ and $\mathbf{V}$, respectively. $\Sigma_r$ is the submatrix formed by the first $r$ rows and first $r$ columns of $\mathbf{\Sigma}_r$. If the rank of matrix $\mathbf{A}$ is much smaller than the number of rows and columns $r m, n$, we will gain a lot in terms of storage.

### Truncated SVD

Recall that in a matrix $\mathbf{\Sigma}$, the diagonal values ​​are non-negative and decreasing $\sigma_1 \geq \sigma_2 \geq \dots \geq \sigma_r \geq 0 = 0 = \dots = 0$. Typically, only a small number of $\sigma_i$ have large values, the remaining values ​​are usually small and close to $0$. Then we can approximate the matrix $\mathbf{A}$ by the sum of $k < r$ rank $1$ matrices:

$$\mathbf{A} \approx \mathbf{A}_k = \mathbf{U}_k \Sigma_k (\mathbf{V}_k)^T = \sigma_1 \mathbf{u}_1 \mathbf{v}^T_1 + \sigma_2\mathbf{u}_2\mathbf{v}_2^2 + \dots + \sigma_k\mathbf{u}_k\mathbf{v}k^T$$

Here is an interesting theorem. It says that the error due to the above approximation is the square root of the sum of the squares of the singular values ​​that we have omitted at the end of $\mathbf{\Sigma}$. Here the error is defined as the Frobineous norm of the difference of the two matrices:

$$||\mathbf{A} - \mathbf{A}_k||_F^2 = \sum_{i = k + 1}^r \sigma_i^2$$

We will prove this theorem as follows: Using the properties of the Frobenius norm, we have $||\mathbf{X}||_F^2 = \text{trace}(\mathbf{X}\mathbf{X}^T)$ and $\text{trace}(\mathbf{XY}) = \text{trace}(\mathbf{YX})$, so we have:

$$
\begin{align*}
    ||\mathbf{A} - \mathbf{A}_k||_F^2 &= \left|\left|\sum_{i = k + 1}^r \sigma_i \mathbf{u}_i\mathbf{v}_i^T \right|\right|_F^2 \newline
    &= \text{trace}\left\{ \left(\sum_{i = k + 1}^r \sigma_i \mathbf{u}_i\mathbf{v}_i^T\right)
    \left(\sum_{j = k + 1}^r \sigma_j \mathbf{u}_j\mathbf{v}_j^T\right)^T
    \right\} \newline
    &= \text{trace}\left\{ \sum_{i = k + 1}^r \sum_{j = k + 1}^r \sigma_i\sigma_j \mathbf{u}_i\mathbf{v}_i^T \mathbf{v}_j \mathbf{u}_j^T
    \right\} \newline
    &= \text{trace}\left\{ \sum_{i = k + 1}^r  \sigma_i^2\mathbf{u}_i\mathbf{u}_i^T
    \right\} \newline
    &= \text{trace}\left\{ \sum_{i = k + 1}^r  \sigma_i^2\mathbf{u}_i^T\mathbf{u}_i \tag{7}
    \right\} \newline
    &= \text{trace}\left\{ \sum_{i = k + 1}^r  \sigma_i^2
    \right\} \newline
    &= \sum_{i = k + 1}^r \sigma_i^2
\end{align*}
$$

The equality $(7)$ is due to $\text{trace}$ function having commutative property. Substitute $k = 0$ and we will have:

$$||\mathbf{A}||_F^2 = \sum_{i = 1}^r \sigma_i^2$$

and

$$\frac{||\mathbf{A} - \mathbf{A}_k||_F^2}{||\mathbf{A}||_F^2} = {\frac{\sum_{i = k + 1}^r \sigma_i^2}{\sum_{j = 1}^r \sigma_j^2}}$$

Thus, the approximation error is smaller if the truncated singular values ​​are smaller than the retained singular values. This is an important theorem that helps determine the matrix approximation based on the amount of information to be retained.

For example, if we want to retain at least 90% of the information in $\mathbf{A}$, we initially calculate $\sum_{j = 1}^r \sigma_j^2$ and then find the smallest $k$ such that 

$$\frac{\sum_{i = 1}^k \sigma_i^2}{\sum_{j = 1}^r \sigma_j^2} \geq 0.9$$

In addition, people has proved that $\mathbf{A}_k$ is the best rank-$k$ approximation of $\mathbf{A}$ in terms of the following optimization problem:

$$\begin{align*}
\min_{\mathbf{B}}||\mathbf{A} - \mathbf{B}||_{F \ \text{or } l_2 \ \text{norm}} \newline
\text{s.t.}\text{rank}(\mathbf{B}) = k
\end{align*}$$

with the definition of norm 2 of a matrix is

$$||\mathbf{A}||_2 = \max_{||\mathbf{x}||_2 = 1} ||\mathbf{Ax}||_2$$

# Application of SVD

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.cluster import KMeans
from numpy import linalg as LA 

img = mpimg.imread('building2.jpg')
plt.imshow(img)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show() 
```

```python
# to gray 
gray = 0.2125 * img[:, :, 0] + 0.7154 * img[:, :, 1] + 0.0721 * img[:, :, 2] #View more: https://en.wikipedia.org/wiki/Grayscale#Luma_coding_in_video_systems 
plt.imshow(gray)
plt.axis('off')
plt.show()
```

```python
U, S, V = LA.svd(gray) 
```

```python
plt.semilogy(S) # semilogy plot: plot y on log scale and x on linear scale
plt.xlabel('$k$', fontsize = 20)
plt.ylabel('$\sigma_k$', fontsize = 20)
# We change the fontsize of minor ticks label 
plt.tick_params(axis='both', labelsize=18)
plt.tight_layout()
plt.show()
```

```python
a = np.sum(S**2)
b = np.zeros_like(S)

for i in range(S.shape[0]):
    b[i] = np.sum(S[:i+1]**2, axis = 0) / a

plt.plot(b)
plt.xlabel('$k$', fontsize = 20)
plt.ylabel('$\\frac{\|\mathbf{A} - \mathbf{A}_k\|_F^2}{\|\mathbf{A}\|_F^2}$', fontsize = 20)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.tight_layout()
plt.show()

err = 1 - b
print('Error =', sum(err))
```

```python
def approx_rank_k(U, S, V, k):
    Uk = U[:, :k]
    Sk = S[:k]
    Vk = V[:k, :]
    return np.around(Uk.dot(np.diag(Sk)).dot(Vk))

Ak = []
K = [5, 50, 100]
for k in K:
    Ak.append(approx_rank_k(U, S, V, k))
```

```python
nrows, ncols = 1, 3
width = 4 * ncols
height = 4 * nrows

fig, axs = plt.subplots(nrows, ncols, figsize=(width, height)) 
for i in range(3):
    axs[i].imshow(Ak[i], cmap = 'gray')
    axs[i].axis('off')
    label = '$k$ = %d: error = %.4f' % (K[i], err[K[i]])
    axs[i].set_title(label, fontsize = 15)
```

```python
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation 

fig, ax = plt.subplots()
def update(i):
    ani = plt.imshow(Ak[i], cmap = 'gray')
    label = '$k$ = %d: error = %.4f' % (5*i + 5, err[i])
    ax.set_xlabel(label)
    ani = plt.axis('off')
    ani = plt.title(label)
    return ani, ax 

anim = FuncAnimation(fig, update, frames=np.arange(0, len(Ak)), interval=500)
anim.save('svd.gif', dpi = 300, writer = 'imagemagick')
# `dpi` is the resolution of the figure. The default is 300.
# `writer` is the program used to write the animation
# `imagemagick` is a popular program to write animations
plt.show()
```
